```python
import pandas as pd
import huggingface_hub
```

This downloads our public dataset which contains mostly document metadata (extracted plain text is also there but isn't very useful as-is).

```python
path = huggingface_hub.hf_hub_download(
    repo_id="demokratis/consultation-documents",
    filename="consultation-documents-preprocessed.parquet",
    repo_type="dataset",
)
df_full = pd.read_parquet(path)
```

This is a reasonable filter surfacing recent law drafts on the federal level:

```python
df = df_full.loc[
    (df_full["document_language"] == "de")  # "fr" and "it" are also available
    & (df_full["political_body"] == "ch")  # filter for federal documents
    & (df_full["document_type"] == "DRAFT")  # filter for the proposed text of the law
    & (df_full["consultation_start_date"].dt.year >= 2025)  # look at recent documents only
]
df.head()
```

This table makes it easy to view the actual PDF documents – they're available for download on Fedlex:

```python
df[
    [
        "document_uuid",
        "consultation_identifier",
        "document_publication_date",
        "consultation_title",
        "document_source_url",
    ]
].sort_values("document_publication_date", ascending=False).style.format(
    {"document_source_url": lambda url: f'<a href="{url}" target="_blank">{url}</a>'}
).set_properties(**{"text-align": "left"})
```
